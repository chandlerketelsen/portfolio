# Density Functional Theory: Hohenberg-Kohn Theorem

Lesson and proof for Hohenberg-Kohn Theorem derived from École Polytechnique.

## Uniqueness of Functionals for QM Observations

Uniqueness arguments will be made for functionals describing quantum mechanical observations

$$ \langle \Psi^0 | \hat{O} | \Psi^0 \rangle = O[n]. $$

Indeed, it can be demonstrated that every observable in the ground state without degeneracies is a unique functional of the electron density $n$. Given the electron density $n$, we have one and only one value for our observables (i.e. measurements). 

Hohenberg-Kohn theorem was proven by P. Hohenberg during his sabattical in Paris, France. The theorem had been conjectured long before the 1960s, however this is the decade when Hohenberg gave concrete mathematical foundation for the theorem. 

## Meaning of Hohenberg-Kohn Theorem

Suppose that each observable is a functional of the many body wavefunction

$$ O = \langle \Psi | \hat{O} | \Psi \rangle = O[\Psi]. $$

Even though observables are simple functionals of the many-body wavefunction, the many-body wavefunction represents the complexity of the system. So for very complex molecules determining the entire wavefunction $\Psi$ can be insurmountable.

### The Many-Body Hamiltonian

In the non-relativistic many-body Hamiltonian function (Born-Oppenheimer) 

$$ H = T + V_{ee} + V_{ext}, $$ 

which term describes the system?

Kinetic Energy and electron-electron Coulomb interactions are the same in each system, only the **external potential $V_{ext}$ distinguishes any particular molecule from another**. 

Given the external potential $V_{ext}(\mathbf{r})$ we know the Hamiltonian 

$$V_{ext}(\mathbf{r}) \rightarrow H $$

and the wavefunction $\Psi$ and eigenvalues $E$ are determined by the hamiltonian

$$ H \rightarrow \Psi, E. $$

Thus, given the external potential $V_{ext}$ we can determine the expectation value of any other quantity like the electron density function $n(\mathbf{r})$: 

$$ n(\mathbf{r}) = \langle \Psi | \sum_i \delta(\mathbf{r} - \mathbf{r}_i) | \Psi \rangle $$

which is the expectation value of the density operator. 

### Motivation

If the expectation value of an operator depends on the wavefunction, in turn depending on the Hamiltonian or the external potential, can we say each observable is uniquely determined by external potential $V_{ext}(\mathbf{r})$ for the ground state of the system? In other words, is each observable a unique functional of the external potential $V_{ext}(\mathbf{r})$?

Uniqueness holds only if $V_{ext}$ and $\Psi$ have one-to-one corresponsence which is a mathematical bijection.

Given external potential $V_{ext}(\mathbf{r})$, we compute the Hamitonian $H$ then diagonalize to obtain the components of wavefunction $\Psi$. Neglect degeneracies for simplicity however note that generalizations are possible. 

The argument will be extended from mappings $V_{ext} \rightarrow \Psi$ to mappings $V_{ext} \rightarrow n(\mathbf{r})$. 


## Character of Diagonalization Mapping

To determine whether the mappings $V_{ext} \rightarrow \Psi$ and $V_{ext} \rightarrow n(\mathbf{r})$ are one-to-one (bijections) consider the case that two different extermal potentials $V_{ext}^{(1)}$, $V_{ext}^{(2)}$ map to the same wavefunction $\Psi$. We compare energy levels of the eigenfunctions to demonstrate *reductio ad absurdum* starting from the non-injective assumption. 

### Uniqueness of Wavefunction: Reductio ad Absurdum

Start by asserting that $V_{ext}^{(1)}$, $V_{ext}^{(2)}$ differ by more than a simple constant 

$$ V_{ext}^{(1)}(\mathbf{r}) \neq V_{ext}^{(2)}(\mathbf{r}) + \text{const.} $$

Potential 1 $V_{ext}^{(1)}$ gives the Hamiltonian 

$$ H_1 = T + V_{ee} + V_{ext}^{(1)} \rightarrow H_1 |\psi\rangle = E_1 |\psi\rangle $$

and Potential 2 $V_{ext}^{(2)}$ gives the other Hamiltonian

$$ H_2 = T + V_{ee} + V_{ext}^{(2)} \rightarrow H_2 |\psi\rangle = E_2 |\psi\rangle $$

where ground state energy levels $E_1$, $E_2$ are found by diagonalizing the Hamiltonian. 

The initial assumption gives that two different potentials we have the same ground-state wavefunction $\psi$. Notice that differencing the two Schrödinger equations $H_1 |\psi\rangle = E_1 |\psi\rangle$ and $ H_2 |\psi\rangle = E_2 |\psi\rangle $ results in the equation 

$$ V_{ext}^{(1)}(\mathbf{r}) = V_{ext}^{(2)}(\mathbf{r}) + (E_1 - E_2), $$

a condtradiction to the initial assumption that two different external potentials lead to the same ground-state wavefunction. Thus, 

$$V_{ext} \leftrightarrow \Psi. \hspace{2cm} \blacksquare$$

### Observables

Since each observable uniquely comes from the wavefunction via expectation value, every observable is also a unique functional of the external potential

$$ O = \langle \Psi | \hat{O} | \Psi \rangle = O[V_{ext}]. $$

Next, what is the relatioship between external potential and the electon density of the system? Note the mappings connect together like

$$ V_{ext} \leftrightarrow \Psi \rightarrow n(\mathbf{r}). $$

Electron density is automatically determined to be a function, since it's determined by integration ofsquare modulus of the wavefunction 

$$ n(\mathbf{r}) = N \int d\mathbf{r}_2 \cdots d\mathbf{r}_n |\Psi(\mathbf{r}_1, \mathbf{r}_2, \cdots, \mathbf{r}_n) |^2 $$

and thus the relationship between $V_{ext}$ and $n(\mathbf{r})$ is *at least* a surjective mapping. Is it also a bijective function? If so, the external potential unqiely determines the system. 

### Uniqueness of Electron Density: Reductio ad Absurdum

With the first potential we have the first Schrödinger equation

$$ H_1 = T + V_{ee} + V_{ext}^{(1)} \rightarrow H_1 |\psi_1\rangle = E_1 |\psi_1\rangle $$

which leads to the eigenvalue $E_1$ and eigenfunction $\psi_1$. Similarly,

$$ H_2 = T + V_{ee} + V_{ext}^{(2)} \rightarrow H_2 |\psi_2\rangle = E_2 |\psi_2\rangle $$

leads to the eigenvalue $E_2$ and eigenfunction $\psi_2$.

Suppose initially that $\psi_1 \rightarrow n(\mathbf{r})$ and $\psi_2 \rightarrow n(\mathbf{r})$ for the same electron density function $n(\mathbf{r})$. 

For the variational theorem, if we replace $\Psi_1$ with any other wavefunction like $\Psi_2$ then the total energy is larger in the latter case. Write

$$ E_1 = \langle \Psi_1 | H_1 | \Psi_1 \rangle < \langle \Psi_2 | H_1 | \Psi_2 \rangle = \langle \Psi_2 | H_1 - H_2 + H_2 | \Psi_2 \rangle $$

where the Hamiltonian $H_2$ has been added and subtracted for no change in the quantity. To proceed, evaluate the energy integrals: 

\begin{align}
E_1 &= \langle \Psi_1 | H_1 | \Psi_1 \rangle \\
    &= \langle \Psi_2 | H_1 - H_2 + H_2 | \Psi_2 \rangle \\
    &= E_2 + \langle \Psi_2 | H_1 - H_2 | \Psi_2 \rangle \\
    &= E_2 + \langle \Psi_2 | (T + V_{int} + V_{ext}^{(1)}) - (T + V_{int} + V_{ext}^{(2)}) | \Psi_2 \rangle \\
    &= E_2 + \langle \Psi_2 | V_{ext}^{(1)} - V_{ext}^{(2)} | \Psi_2 \rangle \\
    &= E_2 + \int d\mathbf{r}_1 \cdots d\mathbf{r}_N \Psi_2^*(\mathbf{r}_1, \cdots, \mathbf{r}_N) [v_{ext}^{(1)}(\mathbf{r}_1) - v_{ext}^{(2)}(\mathbf{r}_1)] \Psi_2(\mathbf{r}_1, \cdots, \mathbf{r}_N) \\
    &= E_2 + \int d\mathbf{r}_1 [v_{ext}^{(1)}(\mathbf{r}_1) - v_{ext}^{(2)}(\mathbf{r}_1)] \int d\mathbf{r}_2 \cdots d\mathbf{r}_N |\Psi_2(\mathbf{r}_1, \mathbf{r}_2, \cdots, \mathbf{r}_N)|^2 \\
    &= E_2 + \int d\mathbf{r} [v_{ext}^{(1)}(\mathbf{r}) - v_{ext}^{(2)}(\mathbf{r})] n_2(\mathbf{r})
\end{align}

$$ \therefore \hspace{1.5cm} E_1 < E_2 + \int d\mathbf{r} \left[ v_{ext}^{(1)} - v_{ext}^{(2)} \right] n_2(\mathbf{r}). $$

Meanwhile, by a similar argument and the same procedure,

$$ \therefore \hspace{1.5cm} E_2 < E_1 + \int d\mathbf{r} \left[ v_{ext}^{(2)} - v_{ext}^{(1)} \right] n_1(\mathbf{r}). $$

Summing the two inequalities and setting $n_1(\mathbf{r}) = n_2(\mathbf{r})$ we obtain 

$$ E_1 + E_2 < E_2 + E_1 $$

which is a clear contradiction. Thus, 

$$ \Psi \leftrightarrow n(\mathbf{r}). \hspace{2cm} \blacksquare $$

We can correctly write that 

$$ V_{ext} \leftrightarrow \Psi \leftrightarrow n(\mathbf{r}) $$

and 

$$ \therefore \hspace{1.5cm} V_{ext} \leftrightarrow n(\mathbf{r}), $$

a bijection. In other words, the external potential $V_{ext}$ uniquely determines the distribution of electrons in a molecule or other many-body system.
