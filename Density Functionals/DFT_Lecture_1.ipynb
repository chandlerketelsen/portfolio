## Introduction

This theory yields usable models of many-body quantum mechanical systems. Exact forms for geometric molecules are evaluated using the Hamiltonian formulation for counts of each electron-electron, electron-nucleus, and nucleus-nucleus interaction. 

$$ \hat{\cal{H}} = - \frac{\hbar^2}{2 m_e} \sum_i \nabla_i^2 - \sum_i \frac{Z_I e^2}{|\mathbf{r_i} - \mathbf{R_I}|} + \frac{1}{2} \sum_{i, j \neq i} \frac{e^2}{\mathbf{r}_i - \mathbf{r_j}} - \frac{\hbar^2}{2} \sum_I \frac{\nabla_i^2}{M_I} + \frac{1}{2} \sum_{I, J \neq I} \frac{Z_I Z_J e^2}{|\mathbf{R}_I - \mathbf{R}_J|} $$



To simplify the Hamiltonian operator, take the nuclei KE and cross-nucleus Coulomb interactions (final two terms) as neglegible. Thus

$$ \hat{\cal{H}} = - \frac{\hbar^2}{2 m_e} \sum_i \nabla_i^2 - \sum_i \frac{Z_I e^2}{|\mathbf{r_i} - \mathbf{R_I}|} + \frac{1}{2} \sum_{i, j \neq i} \frac{e^2}{\mathbf{r}_i - \mathbf{r_j}} $$

and we can use atomic units $ m_e = \hbar = e = 1 $ when necessary.


### External Forces

Define the sum of potentials 

$$ \sum_i v_{ext} (\mathbf{r}_i) = - \sum_i \frac{Z_I e^2}{|\mathbf{r}_i - \mathbf{R}_I|} $$

as forces acting on the bodies (electrons) beside the effects of electrons on each other. This amounts to the nuclear Coulomb energy. 


### Many-Body Hamiltonian

For a given molecule, the system of equations containing all electron superposition information is called $\Psi$ and the Many-Body Hamiltonian. We write it like

$$ \hat{\cal{H}} (\mathbf{r}_1, \mathbf{r}_2, ... \mathbf{r}_N) \Psi_{\lambda} (x_1, x_2, ..., x_N) = E_{\lambda} \Psi_{\lambda} (x_1, x_2, ..., x_N), \hspace{2cm} x_i \equiv (\mathbf{r}_i, \sigma_i) $$

for the $\lambda$ eigenstate of the molecule. Even for a coarse grid, computing and storing this system of equations on a hard drive would require an astronomical amount of storage space. 

Hartree's approximation lets the wavefunction be written as a product of each body's wavefunction as determined by Shrodinger's equation. 

$$ \Psi_{\lambda} (x_1, x_2, ..., x_N) = \psi_{n_1} (x_1) \psi_{n_2} (x_2) ... \psi_{n_N} (x_N), \hspace{2cm} \lambda = \{n_1, n_2, ..., n_N \} $$

When the Fock approximation is accepted, the Pauli exclusion principle is neglected so more than two electrons can occupy ground state of the Hydrogen atom, the 1S orbital. This is the essence of the Fock approximation and can be improved upon. 



### Pauli Effects

By using the Slater determinant 

$$ \Psi = \frac{1}{\sqrt{N!}} \det \begin{pmatrix} \psi_1 (\mathbf{r_1}, \sigma_1) & \psi_1 (\mathbf{r_2}, \sigma_2) & \psi_1 (\mathbf{r_3}, \sigma_3) \\ \psi_2 (\mathbf{r_1}, \sigma_1) & \psi_2 (\mathbf{r_2}, \sigma_2) & \psi_2 (\mathbf{r_3}, \sigma_3) & \cdots \\ \psi_3 (\mathbf{r_1}, \sigma_1) & \psi_3 (\mathbf{r_2}, \sigma_2) & \psi_3 (\mathbf{r_3}, \sigma_3) \\ & \vdots & & \ddots \end{pmatrix}  $$

as the wavefunction some quantum effects are captured. When the positions $ (\mathbf{r}_i, \sigma_i) $ from two different columns change places in the matrix, the sign must be changed in accordance with antisymmetry. Additionally, Pauli's exclusion principle is obeyed since only one particle can occupy each unique state at a time.



## Computing the Many-Body Hamiltonian

Consider the Grand Canonical Ensemble for the many-body system: 

$$ O = \langle \hat{O} \rangle = \frac{1}{Z} \sum_{\alpha} e^{-\beta (E_{\alpha} - \mu N_{\alpha})} \langle \Psi_{\alpha} | \hat{O} | \Psi_{\alpha} \rangle $$

with $ \beta = \frac{1}{k_B T} $, $\alpha$-state energy $E_\alpha$, chemical potential $\mu$ and particle count $N_\alpha$. 

Explicitly, the partition function $ Z $ is written

$$ Z = \sum_{\alpha} e^{-\beta(E_\alpha - \mu N_{\alpha})} $$

Let $ T \approx 0 $ and fix $ N_{\alpha} = \text{const.} $

$$ O = \int \cdots \int dx_1 \cdots dx_N \left[ \Psi^* (x_1, \cdots, x_N) \sum_{i, j, \cdots} O(x_i, x_j, \cdots) \Psi (x_1, \cdots, x_N) \right] $$

Observable $ O $ is a functional $ O[\Psi] $ of the many-body wavefunction and, by symmetry arguments,  can be simplified into a functional of external potentials $ \sum_i v_{ext} $: 

$$ O = O[\Psi] \approx O[v_{ext}] $$

This form assumes a given electrical KE and electron-electron PE. Remember that $v_{ext}$ amounts to nucleus-electron Coulomb interaction. 

Grand Canonical Ensemble is a mathematical trace, which is an invariant, and thus 

$$ \langle \hat{O} \rangle = \frac{1}{Z} \mathrm{Tr} \left\{ e^{-\beta(\hat{\cal{H}} - \mu \hat{N})} \hat{O}\right\}$$

System dependence is expressed by the Hamiltonian. Now the observable $O$ is written in terms of the external potential applied to the system instead of the $3N$ parameters seen before.


## Examples of Quantum Observables IRL

Keep in mind there is purpose behind this formulation of quantum systems. Observables in real life include: 

* Angles made between nuclei of geometric molecules, i.e. $NH_3$, $CO_2$
* Lattice vibrations that give rise to phonons
* Bulk Modulus: ability for a material to resist external pressure

The Lennard-Jones potential is a rudimentary insight to the observables we compute in DFT. By minimizing the Lennard-Jones potential, equilibrium bond length in the $Ar_2$ molecule is determined. 


```python

# Classical Example of an Observable

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

# Dark background
plt.style.use('dark_background')

# Argon-Argon parameters
epsilon = 0.0103  # eV
sigma = 3.4       # Angstroms

# Lennard-Jones function
def lennard_jones(r):
    return 4 * epsilon * ((sigma / r)**12 - (sigma / r)**6)

# Generate r values around minimum
r = np.linspace(3.0, 8.0, 500)
V = lennard_jones(r)

# Find minimum
result = minimize_scalar(lennard_jones, bounds=(3.0, 8.0), method='bounded')
r_eq = result.x
V_eq = result.fun

# Plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(r, V, color='cyan', label='Lennard-Jones Potential (Ar-Ar)')
ax.scatter(r_eq, V_eq, color='red', s=50, label=f'Equilibrium: {r_eq:.2f} Ã…')
ax.axhline(0, color='white', linestyle='--', alpha=0.5)

ax.set_xlabel('Interatomic Distance (Ã…)', fontsize=12)
ax.set_ylabel('Potential Energy (eV)', fontsize=12)
ax.set_title('Lennard-Jones Potential', fontsize=14)
ax.legend()
ax.grid(True, linestyle='--', alpha=0.5)

plt.show()

print(f"Equilibrium bond length (Arâ€“Ar): {r_eq:.4f} Ã…")
```


### Explanation

The key insight from (classical) LJ potential is that all quantum observables can be evaluated by knowledge of ground-state total energy of the system, wherein atomic position plays a crucial role.
