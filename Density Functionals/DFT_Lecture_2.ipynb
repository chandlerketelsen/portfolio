## Why Density Functionals

Density functionals have useful applications in fields beyond the Electronic Problem: From Superconductivity to Alloys, Nuclear Energy to Geochemistry, Stellar Materials and beyond. This lecture series focuses on the electronic problem to build a foundation in the theory, which will then be applied by the physicist or mathematician to real-world issues as needed. 


## Strategy for Computing Observables

There's a tradeoff between computing physical observables using a complicated wavefunction $ \Psi $ with a simple functional, the expectation value integrals, and using a complicated functional with a simplified wavefunction. 

For example, an observable $ O(x_i) $ is an observable of an electron in the system. We write using functional notation 

$$ O = O[ \Psi ] = \int \cdots \int dx_1 \cdots dx_N \Psi^*(x_1, \cdots, x_N) \sum_{i, j, \cdots} O (x_1, \cdots, x_N) \Psi (x_1, \cdots, x_N). $$

While the form is understandable, it is not usable since we cannot compute and store the wavefunction $ \Psi $. So we attempt to compute a functional which depends on a simplified function and we call it

$$ \tilde{O}[ v_{ext} ] $$

since the system changes only with respect to external forces. 

### One-Body Number Density

Try using a local, one-body functional to write the functional

$$ O_1 = \int \cdots \int dx_1 \cdots dx_N \Psi^*(x_1, \cdots, x_N) \sum_{i} O_1 (x_i) \Psi (x_1, \cdots, x_N). $$

All electrons $i > 1$ are identical to the first electron $i = 1$ so the sum is replaced with a factor of $N$: 

$$ O_1 = N \cdot \int \cdots \int dx_1 \cdots dx_N \Psi^*(x_1, \cdots, x_N) O_1 (x_1) \Psi (x_1, \cdots, x_N) $$

By rearranging the seperable integral we arrive at 

$$ O_1 = \int dx_1 O_1(x_1) \cdot N \int dx_2 \cdots dx_N \Psi^*(x_1, \cdots, x_N)\Psi (x_1, \cdots, x_N) $$

then, by integrating out $x_2, \cdots x_N$, we arrive at a desirable result involving number density $n_1(x_1)$

$$ \boxed{O_1 = \int dx_1 O_1(x_1) \cdot n(x_1)}. $$

### Expectation Two-Body Coulomb Operator 

Coulomb repulsion between two electrons is a repeated interaction in many-body system. Using the technique from above, we can separate the operator

$$ V_{ee} = \int \cdots \int dx_1 \cdots dx_N \Psi^* (x_1, \cdots, x_N) \sum_{i, j > i} \frac{e^2}{|\mathbf{r}_i - \mathbf{r}_j |} \Psi (x_1, \cdots, x_2) $$

by removing electrons without replacement for the factor $N \cdot (N-1) $ to replace the sum $ \sum_{i, j > i} $. The functional is written

$$ V_{ee} = \int dx_1 dx_2 \frac{1}{2} \frac{e^2}{ |\mathbf{r}_i - \mathbf{r}_j | } \cdot N(N-1) \int dx_3 \cdots dx_N \Psi^* (x_3, \cdots, x_N) \Psi (x_3, \cdots, x_N) $$

which has separated the 2-body coulomb operator from number density: 

$$ \boxed{V_{ee} = \int dx_1 dx_2 \frac{1}{2} \frac{e^2}{ |\mathbf{r}_i - \mathbf{r}_j | } \cdot n^{(2)} (x_1, x_2)}. $$

### Expectation KE Operator

The electronic KE operator $ O = \sum_i \frac{-\nabla_i^2}{2} $ is integrated with the wavefunction to obtain

$$ E_{K} = - \int \cdots \int dx_1 \cdots dx_N \Psi^*(x_1, \cdots, x_N) \sum_i \frac{\nabla_i^2}{2} \Psi(x_1, \cdots, x_N). $$

To arrive at the result, we only differentiate the wavefuntion on the right by introducing dummy variable $x'_1$ and its operator $\nabla'_1$. At the end of the computation, make it equal to $x_1$ again. Using the identical interactions argument, the sum is replaced to obtain

$$ E_K = - \int dx_1 \left[ \frac{\nabla'^2_1}{2} N \int \cdots \int dx_2 \cdots dx_N \Psi^*(x_1, \cdots, x_N) \Psi(x'_1, \cdots, x_N) \right]_{|x'_1 \rightarrow x_1} $$

which contains the one-body density matrix $\rho$: 

$$ \boxed{ E_K =  -\int dx_1 \left[ \frac{\nabla'^2_1}{2} \rho(x_1, x'_1) \right] } .$$



## Conclusion

### Obstacle: Why we Cannot Proceed

Quantities $Q$ like the one-body number density, two-body number density, and one-body density matrix still **cannot be calculated** since we **cannot store the many-body wavefunction** $\Psi$. 

### Solution: Choose a Different Problem

To avoid the need for data in $\Psi$ we admit a different $Q_i$ for every $O_i[Q_i]$. That means we dont *need* the data $\Psi$ necessarily so long as quantities $Q$ can be measured another way. 

To proceed, pose the problem as follows: 

* Find a quantity $Q$ which makes $O_i[Q]$ computable
* Compute $O_i[Q]$ for observables we want
* Find a way to calculate the $Q$ which functionals $O_i[Q]$ have in common. 

So now, instead of representing the observable as a functional of the many-body wavefunction

$$ O = O[\Psi] $$

we can represent it as a functional of the single-body number density

$$ \cal{O} = \cal{O}[n_1] $$
